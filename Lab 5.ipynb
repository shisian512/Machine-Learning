```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler')
    
])
```

```python
from sklearn.compose import ColumnTransformer

num_feats = list(house_num)
cat_feats = ['ocean_proximity']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_feats),
    ('cat', OneHotEncoder(), cat_feats)
])
```

```python
house_prepared = full_pipeline.fit_transform(house)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(house_prepared, house_label) # perform training
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

pred = lr.predict(house_prepared)
mse = mean_squared_error(house_label, pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(house_label,pred)

print("MSE : ", mse)
print("RMSE : ", rmse)
print("MAE : ", mae)
```
